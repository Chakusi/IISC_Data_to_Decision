```python
import matplotlib.pyplot as plt
import pandas as pandas
import numpy as np

# Step 1: Read the CSV file
data = pandas.read_csv('../data/City_wise_raw_data_1Day_2022_Delhi_1Day.csv')

# Step 2: Extract data for plotting

days = data['Timestamp']
variables = data.columns[1:10]

# Step 3: Plot the data
plt.figure(figsize=(10, 6))

for var in variables:
    plt.plot(days, data[var], marker='o', label=var)

# Step 4: Customize the plot
plt.title('Data Visualization of Delhi (PM2.5, PM10, NO, SO, O, CO) for 2022.', fontsize=14)
plt.xlabel('Days 2022', fontsize=12)
plt.ylabel('Values g/m', fontsize=12)
plt.legend(title='Variables', loc='upper left')
plt.grid(True)
plt.tight_layout()

# Step 5: Show the plot
plt.show()
```
